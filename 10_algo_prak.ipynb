```python
print('@@@  @@@@ @@@    @@    @@@ @@@@')
print('@  @ @  @  @@@   @@   @@@  @  @')
print('@@@@ @  @   @@@ @  @ @@@   @  @')
print('@  @ @  @    @@@    @@@    @  @')
print('@@@  @@@@     @@    @@     @@@@')
mahasiswa = []

def data_masuk():
    nama = input("Masukkan Nama Mahasiswa: ")
    nilai1 = int(input("Masukkan Nilai Praktikum 1: "))
    nilai2 = int(input("Masukkan Nilai Praktikum 2: "))
    nilai3 = int(input("Masukkan Nilai Praktikum 3: "))
    data = {'Nama': nama, 'Praktikum': [nilai1, nilai2, nilai3]}
    mahasiswa.append(data)
    main()

def data():
    print(" NAMA |", "PRAK 1|", "PRAK 2|", "PRAK 3|")
    for i in mahasiswa:
        print(i['Nama'], "     ", i['Praktikum'][0], "     ", i['Praktikum'][1], "     ", i['Praktikum'][2])
    main()

def nilaimahasiswa():
    for i in mahasiswa:
        rata_rata = sum(i['Praktikum']) / len(i['Praktikum'])
        print(f"Rata-rata praktikum {i['Nama']}: {rata_rata:.2f}")
    main()

def nilaipraktikum():
    praktikum1 = sum([m['Praktikum'][0] for m in mahasiswa]) / len(mahasiswa)
    praktikum2 = sum([m['Praktikum'][1] for m in mahasiswa]) / len(mahasiswa)
    praktikum3 = sum([m['Praktikum'][2] for m in mahasiswa]) / len(mahasiswa)
    print("Nilai rata-rata 1= ", praktikum1)
    print("Nilai rata-rata 2= ", praktikum2)
    print("Nilai rata-rata 3= ", praktikum3)
    main()

def nilairatarata():
    nama_cari = input("Masukkan Nama Mahasiswa yang ingin dicari rata-ratanya: ")
    found = False
    for i in mahasiswa:
        if i['Nama'] == nama_cari:
            total_praktikum = sum(i['Praktikum'])
            rata_rata_praktikum = total_praktikum / len(i['Praktikum'])
            print(f"Rata-rata praktikum {i['Nama']}: {rata_rata_praktikum:.2f}")
            found = True
            break

    if not found:
        print("Mahasiswa dengan nama", nama_cari, "tidak ditemukan.")
    main()

def update_nilai():
    update = input("Nama Yang Dicari: ")
    try:
        nilai = int(input("Ingin Update nilai Praktikum ke-: "))
        if 1 <= nilai <= 3:
            nilai_baru = int(input("Nilai Baru: "))
            for i in mahasiswa:
                if i['Nama'] == update:
                    i['Praktikum'][nilai - 1] = nilai_baru
            main()
        else:
            print("Indeks nilai praktikum harus antara 1 dan 3.")
    except ValueError:
        print("Input nilai praktikum dan nilai baru harus berupa angka.")

def main():
    print("Program Membuat List")
    print("1. Memasukkan nama dan nilai")
    print("2. Lihat data")
    print("3. Nilai rata-rata mahasiswa")
    print("4. Nilai rata-rata setiap praktikum mahasiswa")
    print("5. Mencari nilai rata-rata dari setiap mahasiswa")
    print("6. Mengubah nilai mahasiswa")
    print("7. Keluar")
    pilih = int(input("Silahkan pilih (ketik 7 jika ingin keluar):"))
    if pilih == 1:
        data_masuk()
    elif pilih == 2:
        data()
    elif pilih == 3:
        nilaimahasiswa()
    elif pilih == 4:
        nilaipraktikum()
    elif pilih == 5:
        nilairatarata()
    elif pilih == 6:
        update_nilai()
    elif pilih == 7:
        exit()
    else:
        print("Keyword yang kamu masukkan salah!")
        pilih = int(input("Silahkan pilih (ketik 7 jika ingin keluar):"))

main()
```
